```python

import matplotlib.pyplot as plt
import seaborn as sns

# Boxplot to identify outliers in the 'Fare' column
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot')
plt.show()

# Optional: Remove outliers from 'Fare' using the IQR method
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1
df = df[(df['Fare'] >= Q1 - 1.5 * IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]
```

```python
# Boxplot to visualize outliers
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot')
plt.show()

# Set a custom fare range
min_fare = 10
max_fare = 300

df = df[(df['Fare'] >= min_fare) & (df['Fare'] <= max_fare)]

# Display summary after filtering
print(df['Fare'].describe())
```

```python
# Calculate IQR
Q1 = df['Fare'].quantile(0.25)
Q3 = df['Fare'].quantile(0.75)
IQR = Q3 - Q1

# Remove outliers using IQR
df = df[(df['Fare'] >= Q1 - 1.5 * IQR) & (df['Fare'] <= Q3 + 1.5 * IQR)]

# Display summary
print(df['Fare'].describe())
```

```python

# Boxplot to visualize outliers
plt.figure(figsize=(8, 5))
sns.boxplot(x=df['Fare'])
plt.title('Fare Boxplot (Before Removing Outliers)')
plt.show()

# Scatter plot to see outliers clearly
plt.figure(figsize=(8, 5))
sns.scatterplot(x=df.index, y=df['Fare'])
plt.title('Fare Scatter Plot (Before Removing Outliers)')
plt.xlabel("Index")
plt.ylabel("Fare")
plt.show()
```

```python
# Bar chart for the 'Survived' column
df['Survived'].value_counts().plot(kind='bar', color=['salmon', 'skyblue'])
plt.title('Survival Count')
plt.xlabel('Survived (0 = No, 1 = Yes)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Histogram for the 'Age' column
df['Age'].plot(kind='hist', bins=30, color='lightgreen', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# Compute the correlation matrix
#Youll see that Pclass 1 has high fares and Pclass 3 has low fares,
# confirming the negative correlation.
correlation = df[['Fare', 'Pclass']].corr()
print(correlation)
sns.scatterplot(x=df["Pclass"], y=df["Fare"])
plt.title("Fare vs Pclass")
plt.xlabel("Pclass")
plt.ylabel("Fare")
plt.show()
```
